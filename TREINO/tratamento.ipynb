```python
import pandas as pd
import numpy as np
import re

df = pd.read_excel("Orcamento.xlsx")

display(df)
```

```python
df["Data"] = pd.to_datetime(df["Data"], unit='ms')
df["Data"] = df["Data"].dt.strftime('%d/%m/%Y')
display(df)
```

```python
df["Valor"] = df["Valor"].astype(float)

df["Valor"] = df["Valor"].map("{:,.2f}".format).str.replace(",", "X").str.replace(".", ",").str.replace("X", ".")

print(df["Valor"])
```

```python
df_sem_nulos = df.dropna()
print(df_sem_nulos)
```

```python
df.to_excel('Orcamento_atualizado.xlsx', index=False)
display(df)
```
