```python
import pandas as pd
import numpy as np
import re

df = pd.read_excel("Despesas.xlsx")

display(df)
```

```python
df["Valor_previsto"] = df["Valor_previsto"].astype(float)

df["Valor_previsto"] = df["Valor_previsto"].map("{:,.2f}".format).str.replace(",", "X").str.replace(".", ",").str.replace("X", ".")

print(df["Valor_previsto"])
```

```python
df["Valor_realizado"] = df["Valor_realizado"].astype(float)

df["Valor_realizado"] = df["Valor_realizado"].map("{:,.2f}".format).str.replace(",", "X").str.replace(".", ",").str.replace("X", ".")

print(df["Valor_realizado"])
```

```python
df.to_excel('Despesas_atualizado.xlsx', index=False)
display(df)
```
